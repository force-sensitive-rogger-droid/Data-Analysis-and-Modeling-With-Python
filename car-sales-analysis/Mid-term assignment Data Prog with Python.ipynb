## Question 6 Produce a scatter plot of price vs mileage, coloured by fuel type.


```python
#Creating a seaborn scatterplot. 
#X-axis displayes the Price
#Y-axis displayes the Mileage
#The different colors tell us the fuel type: purple for 'Petrol', lime for 'Diesel', red 'Hybrid'.
myplot=sns.scatterplot(
    data=df,
    x="Price",
    y="Mileage",
    hue="Fuel type", #legend based on the fuel type
    palette={'Petrol': 'Purple', 'Diesel': 'lime', 'Hybrid': 'red'}, #specifying the color for each case of fuel type.
    s=10
)
myplot.set(title = "Car Price vs Mileage, by fuel type", ylabel="Mileage", xlabel="Price")
```


## Question 7 Compute price per kilometre


## Question 8 Plot car price vs year of manufacture, colouring points by manufacturer


```python
#Creating a seaborn scatterplot. 
#X-axis displayes the Price
#Y-axis displayes the Year of manufacture
#The different colors tell us the Manufacturer: purple for 'Ford', gold for 'Porsche', red for 'Toyota', lime for 'VW' and blue for 'BMW'.
myplot=sns.scatterplot(
    data=df,
    x="Price",
    y="Year of manufacture",
    hue="Manufacturer", #legend based on fuel type
    palette={'Ford': 'Purple', 'VW': 'lime', 'Toyota': 'red', 'BMW': 'navy', 'Porsche': 'gold'}, #specifying the color for each case of "Manufacturer".
    s=20,
)
myplot.set(title = "Car Price vs Year of Manufacture, by Manufacturer", ylabel="Year of Manufacture", xlabel="Price")
```


## Question 9 Compare Toyota and BMW.


We will first visualize some data, in order to compare the two Manufacturers.


```python
#Create a data frame only with the manufacturers "BMW" and "Toyota"
comparison = df[(df["Manufacturer"] == "BMW") | (df["Manufacturer"] == "Toyota")]
#Creating a seaborn scatterplot to compare the two manufacturers. 
#X-axis displays the Price
#Y-axis displays the Year of manufacture
#The different colors tell us the Manufacturer: Red for 'Toyota' and blue for 'BMW'.
myplot=sns.scatterplot(
    data=comparison,
    x="Price",
    y="Year of manufacture",
    hue="Manufacturer", #legend based on fuel type
    palette={'Toyota': 'red', 'BMW': 'navy'}, #specifying the color for each case of "Manufacturer".
    s=20,
)
myplot.set(title = "Car Price vs Year of Manufacture, by Manufacturer", ylabel="Year of Manufacture", xlabel="Price")
```


From that scatter plot, it is clear that BMW even though the prices may vary, `BMW` sell the most expensive cars. The difference between prices between the two manufacturers becomes more distinguisable as the years pass by, with `BMW` having a higher price increasing rate than `Toyota`. Both manufacturers produce cars all the years and steadily increase the number of cars as the years pass by, with `Toyota` slightly producing more cars by 2020.


```python
#Creating a seaborn scatterplot to compare the two manufacturers. 
#X-axis displayes the Engine size
#Y-axis displayes the Mileage
#The different colors tell us the Manufacturer: Red for 'Toyota' and blue for 'BMW'.
myplot=sns.scatterplot(
    data=comparison,
    x="Engine size",
    y="Mileage",
    hue="Manufacturer", #legend based on fuel type
    palette={'Toyota': 'red', 'BMW': 'navy'}, #specifying the color for each case of "Manufacturer".
    s=20,
)
myplot.set(title = "Engine size vs Mileage, by Manufacturer", ylabel="Mileage", xlabel="Engine size")
```

The cars from both `manufacturers` have similar mileage. Great differences can be spotted when it comes to engine size. `Toyota` produces smaller engines with size from 1 to approximately 2.5. On the other hand, `BMW` produces engines from size 2 to size 5. Both companies produce engines of size from 2 to a little less than 2.5.
